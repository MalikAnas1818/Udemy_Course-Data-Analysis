# 7. Find Out Number of Courses Per Subjects


```python
df.columns
```

```python
df.head(5)
```

```python
df['subject'].unique()
```

```python
df['subject'].value_counts()
```

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x='subject',data=df)
plt.xticks(rotation=60)
plt.xlabel("Subject Name")
plt.ylabel("Total Number of Subject")
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), int(p.get_height()), 
                ha='center', va='bottom')
plt.show()
```

```python
x=df['subject'].value_counts()
```

```python
labels = [f'{label}: {count}' for label, count in zip(x.index, x)]
plt.pie(x,labels=labels,autopct='%0.1f%%')
plt.legend(bbox_to_anchor =(2,1))
#plt.legend()
plt.show()
```
